```python
import glob

noisy_sounds = glob.glob('/Volumes/MacPema/Stuttgart/Sem1_Courses/SpeechSignalProcessing(6)/Proj/Noise-Suppression-Using-Deep-Learning/dataset/NoisyData/*')
```

```python
# ORIGINAL NOISY SOUND
from IPython.display import Audio

audio_path = noisy_sounds[4]
Audio(audio_path,rate=16000)
```

# noisereduce

```python
from scipy.io import wavfile
import noisereduce as nr


rate, data = wavfile.read(audio_path)
reduced_noise = nr.reduce_noise(y=data, sr=rate)

Audio(reduced_noise,rate=16000)
```
